## Problem 5
Use the Pokemon dataset to create a clustermap with color. First, filter the dataframe to only keep data with `Type 1` equals to following values: `Water`, `Normal`, `Grass`, `Bug` and `Psychic`. Annotate the dendrogram using different colors for these five different `Type 1` values. Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv

```python
# Problem 5
# Your code here . . .
"""""
Problem 5
Use the Pokemon dataset to create a clustermap with color.
First, filter the dataframe to only keep data with `Type 1` equals to following values: 
`Water`, `Normal`, `Grass`, `Bug` and `Psychic`. 
Annotate the dendrogram using different colors for these five different `Type 1` values.
Use `Name` as the index.

pokemon_data is in https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv
"""
pokemon_url = "https://raw.githubusercontent.com/csbfx/advpy122-data/master/Pokemon.csv"
pokemon = pd.read_csv(pokemon_url)
### First, filter the dataframe to only keep data with `Type 1` equals to following values: 
###`Water`, `Normal`, `Grass`, `Bug` and `Psychic`. 
types_to_keep = ["Water", "Normal", "Grass", "Bug", "Psychic"]
df = pokemon[pokemon["Type 1"].isin(types_to_keep)].copy()
### use name as the index 
df=df.set_index("Name")
## use only the nummeric columns for cluster 
numeric_cols = ["Total", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
pokemon_num = df[numeric_cols]
## use different colors for these five different Type 1 values
type_colors = {
    "Water": "blue",
    "Normal": "gray",
    "Grass": "green",
    "Bug": "yellow",
    "Psychic": "purple"
}
row_colors = df["Type 1"].map(type_colors)

sns.clustermap(
    pokemon_num,
    row_colors=row_colors,
    cmap="viridis",
    figsize=(10, 10),
    col_cluster=False
)

```
